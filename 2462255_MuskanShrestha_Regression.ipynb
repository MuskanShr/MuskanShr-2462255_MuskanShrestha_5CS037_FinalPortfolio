Data Cleaning

```python
#  Drop columns with too many missing values (>50% missing = not useful)
missing_pct = dataset.isnull().sum() / len(dataset) * 100
cols_to_remove = missing_pct[missing_pct > 50].index.tolist()
print(f"Dropping {len(cols_to_remove)} columns with >50% missing values")
dataset = dataset.drop(columns=cols_to_remove)

# Fill remaining missing numerical values with median
numerical_features = dataset.select_dtypes(include=['int64', 'float64']).columns.tolist()

for col in numerical_features:
    dataset[col] = dataset[col].fillna(dataset[col].median())

# Drop rows where target is missing
dataset = dataset.dropna(subset=['co2'])

print(f"Final dataset: {dataset.shape[0]} rows, {dataset.shape[1]} columns")
print(f"Missing values remaining: {dataset.isnull().sum().sum()}")
```

```python
target= 'co2'
print(f"Target variable: {target}")
print(f"Target missing: {dataset[target].isnull().sum()}")

df_clean = dataset[dataset[target].notna()].copy()

duplicates = df_clean.duplicated().sum()
print(f"Duplicate rows: {duplicates}")
```

Outlier detection

```python
Q1 = df_clean[target].quantile(0.25)
Q3 = df_clean[target].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df_clean[(df_clean[target] < lower_bound) | (df_clean[target] > upper_bound)]

print(f"Outlier Analysis:")
print(f"Q1: {Q1:.2f}")
print(f"Q3: {Q3:.2f}")
print(f"IQR: {IQR:.2f}")
print(f"Lower bound: {lower_bound:.2f}")
print(f"Upper bound: {upper_bound:.2f}")
print(f"Number of outliers: {len(outliers):,} ({len(outliers)/len(df_clean)*100:.2f}%)")
print("\nDecision: Keeping outliers (they represent real high-emission countries)")

plt.figure(figsize=(6, 4))
plt.boxplot(df_clean[target], vert=False, showfliers=True)
plt.title(f"Outlier Detection using IQR for {target}")
plt.xlabel(target)
plt.tight_layout()
plt.show()
```

 Build Primary Machine Learning Models (Two Classical ML Models):

Train test split

```python
# defining features and target value
X=dataset.drop(columns=['co2','iso_code','country'])
Y=dataset['co2']

# print("Features:\n",X.head(3))
# print("\nTarget:\n",Y.head(3))
```

```python
import sklearn.model_selection as model_selection
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.2, random_state=42)
print("Training set shape:", X_train.shape)
print("Test set shape:", X_test.shape)
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
x_scaler = StandardScaler()
X_train_scaled = x_scaler.fit_transform(X_train)
X_test_scaled = x_scaler.transform(X_test)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled  = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

Linear regression

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(X_train_scaled, Y_train)
```

```python
# make prediction
Y_train_pred=linear_model.predict(X_train_scaled)
Y_test_pred=linear_model.predict(X_test_scaled)
```

```python
from numpy import test
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score
# calculating MSE on scaled data
print("Linear Regression Results:")
train_mse_lr= mean_squared_error(Y_train,Y_train_pred)
test_mse_lr= mean_squared_error(Y_test,Y_test_pred)
print(f'MSE on train data: {train_mse_lr:.3f}')
print(f'MSE on test data: {test_mse_lr:.4f}')

train_rmse_lr= np.sqrt(mean_squared_error(Y_train,Y_train_pred))
test_rmse_lr= np.sqrt(mean_squared_error(Y_test,Y_test_pred))
print(f'\nRMSE on train data: {train_rmse_lr:.3f}')
print(f'RMSE on test data: {test_rmse_lr:.4f}')

train_r2_lr= r2_score(Y_train,Y_train_pred)
test_r2_lr= r2_score(Y_test,Y_test_pred)
print(f'\nR2 on train data: {train_r2_lr:.3f}')
print(f'R2 on test data: {test_r2_lr:.4f}')

train_mae_lr=mean_absolute_error(Y_train,Y_train_pred)
test_mae_lr= mean_absolute_error(Y_test,Y_test_pred)
print(f'\nMAE on train data: {train_mae_lr:.3f}')
print(f'MAE on test data: {test_mae_lr:.4f}')
```

  Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators=50, random_state=42,n_jobs=-1)
rf_model.fit(X_train, Y_train)
```

```python
# make predictions
Y_train_pred_rf = rf_model.predict(X_train)
Y_test_pred_rf = rf_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score
# calculating MSE on scaled data
print("Random Forest Regression Results:")
train_mse_rf= mean_squared_error(Y_train,Y_train_pred_rf)
test_mse_rf= mean_squared_error(Y_test,Y_test_pred_rf)
print(f'MSE on train data: {train_mse_rf:.3f}')
print(f'MSE on test data: {test_mse_rf:.4f}')

train_rmse_rf= np.sqrt(mean_squared_error(Y_train,Y_train_pred_rf))
test_rmse_rf= np.sqrt(mean_squared_error(Y_test,Y_test_pred_rf))
print(f'\nRMSE on train data: {train_rmse_rf:.3f}')
print(f'RMSE on test data: {test_rmse_rf:.4f}')

train_r2_rf= r2_score(Y_train,Y_train_pred_rf)
test_r2_rf= r2_score(Y_test,Y_test_pred_rf)
print(f'\nR2 on train data: {train_r2_rf:.3f}')
print(f'R2 on test data: {test_r2_rf:.4f}')

train_mae_rf=mean_absolute_error(Y_train,Y_train_pred_rf)
test_mae_rf= mean_absolute_error(Y_test,Y_test_pred_rf)
print(f'\nMAE on train data: {train_mae_rf:.3f}')
print(f'MAE on test data: {test_mae_rf:.4f}')
```

```python
comparison_df = pd.DataFrame({
    'Model': ['Linear Regression', 'Random Forest'],
    'Train_MSE': [train_mse_lr, train_mse_rf],
    'Test_MSE': [test_mse_lr, test_mse_rf],
    'Train_RMSE': [train_rmse_lr, train_rmse_rf],
    'Test_RMSE': [test_rmse_lr, test_rmse_rf],
    'Train_MAE': [test_mae_lr, train_mae_rf],
    'Test_MAE': [test_mae_lr, test_mae_rf],
    'Train_R2': [train_r2_lr, train_r2_rf],
    'Test_R2': [test_r2_lr, test_r2_rf]
})

print("MODEL COMPARISON TABLE")
print(comparison_df.to_string(index=False))
```

```python
print("MODEL SELECTION - WHICH IS BETTER?")

if test_r2_rf > test_r2_lr:
    better_model = "Random Forest"
    r2_diff = test_r2_rf - test_r2_lr
    rmse_diff = test_rmse_lr - test_rmse_rf
    print(f"\n WINNER: {better_model}")
    print(f"\n Random Forest performs BETTER than Linear Regression:")
    print(f"   • R² is {r2_diff:.4f} points higher ({(r2_diff/test_r2_lr)*100:.1f}% improvement)")
    print(f"   • RMSE is {rmse_diff:.2f} lower ({(rmse_diff/test_rmse_lr)*100:.1f}% improvement)")
    print(f"\nPerformance Metrics:")
    print(f"   Random Forest R²: {test_r2_rf:.4f}")
    print(f"   Linear Regression R²: {test_r2_lr:.4f}")
else:
    better_model = "Linear Regression"
    r2_diff = test_r2_lr - test_r2_rf
    rmse_diff = test_rmse_rf - test_rmse_lr
    print(f"\n WINNER: {better_model}")
    print(f"\n Linear Regression performs BETTER than Random Forest:")
    print(f"   • R² is {r2_diff:.4f} points higher ({(r2_diff/test_r2_rf)*100:.1f}% improvement)")
    print(f"   • RMSE is {rmse_diff:.2f} lower ({(rmse_diff/test_rmse_rf)*100:.1f}% improvement)")
    print(f"\n Performance Metrics:")
    print(f"   Linear Regression R²: {test_r2_lr:.4f}")
    print(f"   Random Forest R²: {test_r2_rf:.4f}")
```

HyperParameter tuning for linear regression


```python
# hyperparameter tuning for linear regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
alpha = {'alpha' : [0.01, 0.1, 1, 10, 100]}

# ridge
ridge= Ridge()
ridge_cv= GridSearchCV(ridge, alpha, cv=5, scoring='neg_mean_squared_error')
ridge_cv.fit(X_train_scaled, Y_train)
best_ridge = ridge_cv.best_estimator_
print("Best Ridge alpha:", ridge_cv.best_params_)
print("Best CV MSE:", -ridge_cv.best_score_)
```

Train best model and evaluate on test set

```python
# prediction
ridge_pred = best_ridge.predict(X_test_scaled)

# MSE
ridge_test_mse = mean_squared_error(Y_test, ridge_pred)
print(f"Ridge Test Mse: {ridge_test_mse:.4f}")

# rmse
ridge_test_rmse = np.sqrt(mean_squared_error(Y_test, ridge_pred))
print(f"Ridge Test RMSE: {ridge_test_rmse:.4f}")

# r2
ridge_test_r2 = r2_score(Y_test, ridge_pred)
print(f"Ridge Test R2: {ridge_test_r2:.4f}")

# MAE
ridge_test_mae = mean_absolute_error(Y_test, ridge_pred)
print(f"Ridge Test MAE: {ridge_test_mae:.4f}")
```

Feature Selection for Linear Regression

```python
# RFE Feature Selection for linear regression
from sklearn.feature_selection import RFE
estimator = LinearRegression()
num_features_to_select = 5
rfe = RFE(estimator, n_features_to_select=num_features_to_select)

rfe.fit(X_train_scaled, Y_train)

feature_names= X_train.columns

# Get feature ranking and support
feature_ranking = rfe.ranking_
feature_support = rfe.support_

# Print the results
# print("Feature ranking: ", feature_ranking)
# print("Feature support: ", feature_support)

selected_features_lr = feature_names[feature_support]

X_train_lr_selected = X_train_scaled[selected_features_lr]
X_test_lr_selected  = X_test_scaled[selected_features_lr]

print("RFE RESULTS - TOP 5 FEATURES")
for i, feature in enumerate(selected_features_lr, 1):
    print(f"{i}. {feature}")

# Create a dataframe to store feature names and importance scores
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Ranking': feature_ranking, 'Support': feature_support})

# Sort the dataframe by ranking in ascending order
feature_importance_df = feature_importance_df.sort_values(by='Ranking')


# Create a bar chart to visualize feature rankings
plt.figure(figsize=(12, 6))
plt.bar(range(len(feature_importance_df)), feature_importance_df['Ranking'])
plt.xlabel('Features')
plt.ylabel('Ranking')
plt.title('Feature Rankings')
plt.xticks(range(len(feature_importance_df)), feature_importance_df['Feature'], rotation=90)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Calculating metrics
# MSE
train_mse_rf_final = mean_squared_error(Y_train, Y_train_pred_final_rf)
test_mse_rf_final = mean_squared_error(Y_test, Y_test_pred_final_rf)
print(f'MSE on train data: {train_mse_rf_final:.3f}')
print(f'MSE on test data: {test_mse_rf_final:.4f}')

train_rmse_rf_final= np.sqrt(mean_squared_error(Y_train,Y_train_pred_final_rf))
test_rmse_rf_final= np.sqrt(mean_squared_error(Y_test,Y_test_pred_final_rf))
print(f'\nRMSE on train data: {train_rmse_rf_final:.3f}')
print(f'RMSE on test data: {test_rmse_rf_final:.4f}')

train_r2_rf_final= r2_score(Y_train,Y_train_pred_final_rf)
test_r2_rf_final= r2_score(Y_test,Y_test_pred_final_rf)
print(f'\nR2 on train data: {train_r2_rf_final:.3f}')
print(f'R2 on test data: {test_r2_rf_final:.4f}')

train_mae_rf_final=mean_absolute_error(Y_train,Y_train_pred_final_rf)
test_mae_rf_final= mean_absolute_error(Y_test,Y_test_pred_final_rf)
print(f'\nMAE on train data: {train_mae_rf_final:.3f}')
print(f'MAE on test data: {test_mae_rf_final:.4f}')
```

```python
# check for overfitting
r2_diff_rf = train_r2_rf_final - test_r2_rf_final
print(f"Overfitting Check:")
print(f"   Training R²: {train_r2_rf_final:.4f}")
print(f"   Testing R²:  {test_r2_rf_final:.4f}")
print(f"   Difference:  {r2_diff_rf:.4f}")
if r2_diff_rf < 0.05:
    print("No significant overfitting")
elif r2_diff_rf < 0.10:
    print("Slight overfitting")
else:
    print(" Significant overfitting")
```
